### Only looking at data since April 28th

```python
states_latest = states[states['date']>='2020-04-28']
```

## Plots

### New York

```python
plt.figure(figsize=(8,5))
plt.title("Number of Covid-19 Cases and Deaths Over Time for New York State")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")
plt.plot(states[states['state']=='New York']['date'], states[states['state']=='New York']['cases'])
plt.plot(states[states['state']=='New York']['date'], states[states['state']=='New York']['deaths'])

plt.legend(labels=['cases', 'deaths']);
```

```python
plt.figure(figsize=(10,5))
plt.title("Number of Covid-19 Cases and Deaths for New York State since April 28th")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")
plt.plot(states_latest[states_latest['state']=='New York']['date'], states_latest[states_latest['state']=='New York']['cases'])
plt.plot(states_latest[states_latest['state']=='New York']['date'], states_latest[states_latest['state']=='New York']['deaths'])

plt.legend(labels=['cases', 'deaths']);
```

```python
plt.figure(figsize=(8,5))
plt.title("Number of Covid-19 Cases and Deaths Over Time in Texas")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")
plt.plot(states[states['state']=='Texas']['date'], states[states['state']=='Texas']['cases'])
plt.plot(states[states['state']=='Texas']['date'], states[states['state']=='Texas']['deaths'])

plt.legend(labels=['cases', 'deaths'])
plt.vlines(x = states[(states['state']=='Texas') & (states['date']=='2020-04-30')]['date'],
           ymin = 0, ymax = 35000, colors = 'r', linestyle = 'solid', linewidth=2)
plt.text(states[(states['state']=='Texas') & (states['date']=='2020-03-28')]['date'], 30000, f"Stay at home order lifted", fontsize = 12, c = 'r')
;
```

```python
plt.figure(figsize=(10,5))
plt.title("Number of Covid-19 Cases and Deaths Since April 28th in Texas")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")
plt.plot(states_latest[states_latest['state']=='Texas']['date'], states_latest[states_latest['state']=='Texas']['cases'])
plt.plot(states_latest[states_latest['state']=='Texas']['date'], states_latest[states_latest['state']=='Texas']['deaths'])

plt.legend(labels=['cases', 'deaths']);
```

### New York & Texas

```python
plt.figure(figsize=(8,5))
plt.title("Number of Covid-19 Cases and Deaths for New York State and Texas")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")

for state in ['New York', "Texas"]:
    plt.plot(states[states['state']==state]['date'], states[states['state']==state]['cases']);
    plt.plot(states[states['state']==state]['date'], states[states['state']==state]['deaths']);

plt.legend(['New York Cases', "New York Deaths", 'Texas Cases', 'Texas Deaths']);
```

```python
plt.figure(figsize=(10,10))
plt.title("Number of Covid-19 Cases and Deaths for New York State and Texas Since April 28th")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases and Deaths")

for state in ['New York', "Texas"]:
    plt.plot(states_latest[states_latest['state']==state]['date'], states_latest[states_latest['state']==state]['cases']);
    plt.plot(states_latest[states_latest['state']==state]['date'], states_latest[states_latest['state']==state]['deaths']);

plt.legend(['New York Cases', "New York Deaths", 'Texas Cases', 'Texas Deaths']);
```

### All States

```python
plt.figure(figsize=(12,15.5))
plt.title("Number of Covid-19 Cases By State Since April 28th")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases")

for state in states_latest['state'].unique():
    plt.plot(states_latest[states_latest['state']==state]['date'], states_latest[states_latest['state']==state]['cases']);
    
plt.legend([state for state in states_latest['state'].unique()], loc='upper left');
```

```python
plt.figure(figsize=(12,15.5))
plt.title("Number of Covid-19 Cases Over Time By State")
plt.xlabel("Date")
plt.ylabel("Number of Covid-19 Cases")

for state in states['state'].unique():
    plt.plot(states[states['state']==state]['date'], states[states['state']==state]['cases']);
    
plt.legend([state for state in states['state'].unique()], loc='upper left');
```
