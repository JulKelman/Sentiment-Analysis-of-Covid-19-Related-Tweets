## Combining tweets and covid data 

```python
covid_ny_tx = pd.read_csv('./covid_ny_tx.csv')
```

```python
covid_ny_tx
```

```python
covid_ny_tx['date'] = pd.to_datetime(covid_ny_tx['date'])
```

```python
tweets_cleaned['date'] = [timestamp.date() for timestamp in tweets_cleaned['timestamp']]
```

```python
tweets_cleaned['date'] = pd.to_datetime(tweets_cleaned['date'])
```

```python
ny_tweets = tweets_cleaned[tweets_cleaned['state']=='NY']
```

```python
tx_tweets = tweets_cleaned[tweets_cleaned['state']=='TX']
```

```python
ny_covid = covid_ny_tx[covid_ny_tx['state']=='NY']
```

```python
tx_covid = covid_ny_tx[covid_ny_tx['state']=='TX']
```

```python
covid_ny_tx.dtypes
```

```python
tx_covid.dtypes
```

```python
tx_covid
```

```python
tx_all = pd.merge(left=tx_tweets, right=tx_covid, left_on='date', right_on='date')
```

```python
ny_all = pd.merge(left=ny_tweets, right=ny_covid, left_on='date', right_on='date')
```

```python
tweets_and_covid = pd.concat(objs=[ny_all, tx_all], axis=0)
tweets_and_covid.reset_index(drop = True, inplace=True)
```

```python
tweets_and_covid.head()
```

```python
tweets_and_covid.drop(columns='state_y', inplace=True)
```

```python
tweets_and_covid.rename(columns={'timestamp':'tweet_timestamp', 'state_x':'state'}, inplace=True)
```
