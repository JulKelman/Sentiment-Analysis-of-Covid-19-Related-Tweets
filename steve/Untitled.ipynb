```python
import pandas as pd
```

```python
df = pd.read_csv('../data/april28-may6_full.csv')
```

```python
df.tail()
```

```python
ny_index_list = []

for num in range(0,12909):
    if 'NY' in df['location'][num]:
        ny_index_list.append(num)
        
len(ny_index_list)
```

```python
tx_index_list = []

for num in range(0,12909):
    if 'TX' in df['location'][num]:
        tx_index_list.append(num)
        
len(tx_index_list)
```

```python
ny_df = df.loc[ny_index_list,:]
```

```python
ny_df['city'] = [item.split(',')[0] for item in ny_df['location']]
```

```python
ny_df['state'] = [item.split(',')[1] for item in ny_df['location']]
```

```python
tx_df = df.loc[tx_index_list,:]
```

```python
tx_df['city'] = [item.split(',')[0] for item in tx_df['location']]
tx_df['state'] = [item.split(',')[1] for item in tx_df['location']]
```
